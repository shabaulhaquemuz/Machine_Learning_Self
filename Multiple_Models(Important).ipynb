```python
df=df.drop(columns=["EmployeeCount","Over18","StandardHours","EmployeeNumber"])
```

```python
df.head(2)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
df=pd.get_dummies(df,columns=["BusinessTravel","Department","EducationField","Gender","JobRole","MaritalStatus","OverTime"])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Attrition"]=le.fit_transform(df["Attrition"])
```

```python
x=df.drop(columns="Attrition")
```

```python
y=df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df["Attrition"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_train, y_train)

y_train.value_counts()
y_resampled.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_resampled = scaler.fit_transform(x_resampled)  # fit on balanced training data
x_test = scaler.transform(x_test)                # transform test data

# Why rescaling (Standardization) is important
# 1. Different scales across features

# Example:
# Age ranges from ~18–60
# MonthlyIncome can be in the thousands
# DistanceFromHome is small numbers
# Models that rely on distance or optimization (like Logistic Regression, SVM, KNN, Naive Bayes with Gaussian assumption) can get biased toward large-scale features if we don’t rescale.


# 2. Helps gradient-based models
# Logistic Regression uses gradient descent → faster and more stable when features are on the same scale.

# 3. Tree-based models (DecisionTree, RandomForest)
# These models split based on thresholds, so they don’t care about scaling.
# But since you’re training multiple models, it’s safest to scale once for all (it won’t hurt trees, but helps linear/Naive Bayes).


# ✅ Models that require / benefit from scaling
# (They are sensitive to feature magnitude because they use distances, dot products, or gradient-based optimization)
# Logistic Regression
# Linear Regression (especially with regularization: Ridge, Lasso)
# Support Vector Machines (SVMs)
# K-Nearest Neighbors (KNN)
# Principal Component Analysis (PCA) & other dimensionality reduction methods
# Naive Bayes (GaussianNB) → assumes normally distributed features, scaling helps
# Neural Networks (Deep Learning) → scaling speeds up convergence


# 🚫 Models that do NOT require scaling
# (They split or compare values independently, so magnitude doesn’t matter)
# Decision Tree
# Random Forest
# Gradient Boosting (XGBoost, LightGBM, CatBoost, etc.)
# Naive Bayes (MultinomialNB, BernoulliNB) → work with counts/binary values

# Naive Bayes variant that benefits from scaling
# GaussianNB → assumes features follow a normal distribution.
# If your features are on very different scales (e.g., Age = 20–60 vs. MonthlyIncome = thousands), the Gaussian variance estimates can get skewed.
# Standardizing to mean=0, std=1 often improves performance.

# 🚫 Naive Bayes variants that do NOT require scaling
# MultinomialNB → works with counts (e.g., text classification, word frequencies). Scaling would distort the counts.
# BernoulliNB → works with binary features (0/1). Scaling makes no sense.
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000, random_state=42)
lr.fit(x_resampled,y_resampled)
```

```python
y_pred_new=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_new)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_resampled,y_resampled)
```

```python
y_preddt=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_preddt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_resampled,y_resampled)
```

```python
y_predrf=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predrf)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussianclassifier=GaussianNB()
gaussianclassifier.fit(x_resampled,y_resampled)
```

```python
y_predgnb = gaussianclassifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predgnb)
```
