```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Example data
data = pd.DataFrame({
    'Math': [85, 90, 60, 55, 95],
    'Science': [80, 88, 65, 60, 92],
    'English': [82, 86, 63, 58, 91]
})

# Step 1: Standardize
scaler = StandardScaler()
scaled = scaler.fit_transform(data)

# Step 2: Apply PCA
pca = PCA(n_components=1)  # reduce to 1 component
principal = pca.fit_transform(scaled)

print("Principal Component:\n", principal)
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
# Interpretation:
# Only one principal component (PC1) captures 96% of the total variance.
# We’ve reduced 3 features → 1 without losing much information
```
