```python
#Now we have to perform feature scaling i.e standardization method here and to perform any feature scaling either standardization or normalization,
#it is always recommended to perform train_test_split
```

```python
# | Variable      | Gets Data From                 | Meaning                                                       |
# | ------------- | ------------------------------ | ------------------------------------------------------------- |
# | **`x_train`** | `df.drop("Purchased", axis=1)` | Training part of all **independent features** (input columns) |
# | **`x_test`**  | `df.drop("Purchased", axis=1)` | Testing part of the same **independent features**             |
# | **`y_train`** | `df["Purchased"]`              | Training part of the **target column** (Purchased)            |
# | **`y_test`**  | `df["Purchased"]`              | Testing part of the **target column** (Purchased)             |
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns="Purchased"),df["Purchased"], test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
#Now applying standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)

# ❌ We fit only on x_train,
# ✅ We transform both x_train and x_test
# Because:
# fit() → learns parameters (like mean, std, min, max) from the training data.
# transform() → applies those learned parameters to scale the data.

# Why not fit on x_test?
# Because:
# The test set is unseen data — it simulates new, real-world data.
# If you fit the scaler on test data, you are leaking information from the test set into the training process.
# ➤ This causes data leakage, which makes your model evaluation unreliable.
```

```python
print(x_train.head(2))
```

```python
print(x_train_scaled.sample(2))  #read the error, it means our dataframe is converted into numpy array after fit_transformation

# After standardization (or normalization) using scikit-learn scalers,
# the result is a NumPy array, not a pandas DataFrame.
# So yes — the output of fit_transform() is a NumPy array.
# Because scikit-learn (StandardScaler, MinMaxScaler, etc.) is built to work internally with NumPy arrays, not pandas DataFrames.
# It converts your DataFrame into a NumPy array during processing for efficiency.

# If you want to get back a DataFrame:
# You can convert it back manually with the same column names:
# scaled_df = pd.DataFrame(scaled, columns=df.columns)
```

```python
print(x_train_scaled[:2])  ## first 2 rows of numpy array
# x_train data is actually a dataframe but when standardization is done that is "x_train_scaled" it becomes numpy array 
```

```python
#To solve the above problem we have to convert the numpy array into data frame.. i.e. x_train_scaled into dataframe from numpy array
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
# You standardized your DataFrame and then converted the NumPy array back into a DataFrame:
# The new DataFrame (scaled_df) is still standardized —
# the values do not return to the original ones.

# It’s just the data type that changed back (from NumPy array → pandas DataFrame).
# The scaled (standardized) values remain the same.
```
