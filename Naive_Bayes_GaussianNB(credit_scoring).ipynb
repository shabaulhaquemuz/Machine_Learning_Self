```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("credit_scoring.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=["Gender","Marital Status","Education Level"])
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df["Employment Status"]=oe.fit_transform(df[["Employment Status"]])
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Type of Loan"]=oe.fit_transform(df[["Type of Loan"]])
df.head(2)
```

```python
df.astype(int)
```

```python
x=df.drop(columns="Type of Loan")
y=df["Type of Loan"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
# create scaler object
sc = StandardScaler()
```

```python
x_train_new=sc.fit_transform(x_train)
x_test_new=sc.fit_transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussianclassifier=GaussianNB()
gaussianclassifier.fit(x_train,y_train)
```

```python
y_pred = gaussianclassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred,average="micro")
```
